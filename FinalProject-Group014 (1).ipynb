# Overview

In this study, we examined how exposure to emotionally positive words affects the activation of the Default Mode Network (DMN) and the language network/cortex in healthy adults during a passive listening task. We used an fMRI dataset from Pereira et al. (2018)[1] which consists of brain activations from five participants who passively listened to 627 different sentences, later on categorized into positive, neutral, and negative sentiments through sentiment analysis. By comparing the average fMRI activations in response to each sentiment category and by performing logistic regression as a decoder evaluator (from brain to stimuli), results suggest that neutral words actually show a higher activation in the DMN with respect to both negative and positive words - and negative words reveal only a *slightly* higher activation in the DMN when compared to positive words. For the language cortex, positive sentiment words showed a higher activation in comparison to both neutral and negative stimuli. These results weakly support our hypothesis regarding brain activations in response to positive sentiment valued words in the DMN and more strongly supports our hypothesis about the effect of positive affect words on the brain activations of the language cortex.

## Background & Prior Work

The Default Mode Network (DMN) is increasingly recognized for its involvement in internal modes of cognition such as daydreaming, recall of memories, and self-referential thought processes. This brain network has been implicated in various psychological conditions ranging from depression to neurodegenerative diseases, indicating its critical role in mental health and cognitive functioning. 

The paper from Pereira et al. (2018)[1] studying the above-mentioned large scale brain network (DMN) primarily active during wakeful rest explores the development of a universal brain decoder capable of interpreting semantic meaning from brain activation patterns. The new method creates a semantic space using 300-dimensional GloVe vectors to organize words into 180 meaningful clusters via spectral clustering, enhancing interpretability by excluding infrequent or uninformative words. 

This enables the decoder to interpret the meaning of both words and sentences across diverse and abstract topics. The findings of the study particularly highlight the roles of four major brain networks in semantic processing: the language-selective network, the default mode network (DMN), the task-positive (TP) network, and the visual network. The language-selective network was primarily engaged in mapping linguistic forms to meanings, while the DMN, typically associated with internally oriented thought, also showed involvement in semantic processing. Although active during semantic tasks, the TP network's activation patterns did not align as closely with semantic vectors as the language-selective network, indicating its limited role in directly interpreting semantic content. Similarly, while the visual network was activated, its role was primarily in processing the visual aspects of stimuli, not their semantic meanings.

This paper aligns with our project as it highlights the neural plausibility of decoding complex semantic representations, which is relevant to understanding how emotionally charged words might influence brain activity in the DMN and langueg cortex during language processing tasks, particularly passive listening. 

To further understand the role of the DMN, we considered an article studying the DMN’s involvement in depression in which researchers examined the DMN brain activity of depressed individuals during rest and during goal driven tasks in comparison to controls[2] . Using fMRI brain imaging technique, results indicated that individuals with depression were not able to diminish activity in the DMN during goal relevant tasks, exhibited elevated activity in DMN regions when exposed to negative categorized images compared to controls, and showed heightened activity in the DMN even when asked to reappraise their negative experience. This study suggests that individuals with depression show a greater difficulty with emotional regulation, specifically dealing with one of the experimental conditions, “making positive” a scenario. Consequently, our research focus is aimed towards understanding the relationship between positive sentiment valued words and DMN activations, gaining inspiration from exploring the relationship between depression and the DMN, and in motivation to apply our findings to a potential guideline for treatment of depression.

Additionally, a relevant article related to our research topic was that of Juhnyung Kim et. al.[3] which examines the impact of self-talk, both positive and negative, on brain functional connectivity and cognitive performance during the execution of cognitive tasks. By analyzing functional magnetic resonance imaging (fMRI) data, the study found that negative self-talk, specifically self-criticism, led to significant increases in performance on Raven’s Progressive Matrices, a test measuring fluid intelligence. This improvement was accompanied by substantial changes in brain connectivity across several networks, including the default mode network (DMN), the reward-motivation network, and the central-executive network. On the other hand, positive self-talk, characterized as self-respect, influenced brain connectivity in different regions and was associated with increased confidence. However, unlike negative self-talk, it did not lead to significant improvements in task performance. This suggests that while positive self-talk enhances certain psychological states and executive functions, it does not necessarily translate into better performance under the conditions studied. 

The findings of the study suggest that positive self-talk did not specifically focus on passive listening tasks, but rather demonstrated that positive self-talk modulates the DMN alongside other networks during active cognitive engagement, leading to varied cognitive outcomes based on the emotional context of the self-talk. This could imply that positive emotional stimuli, such as positive words, might similarly influence DMN connectivity and function during passive listening, potentially affecting cognitive and emotional processing in healthy adults. Due to the similarity of experimental conditions and motivations, we might say that this could potentially be an observable pattern in our research, as well.

Regarding the language cortex and in aim of analyzing its response to positive, neutral, and negative words - a study researching language related beta brain activations in the left frontal area was assessed in Major Depression (MDD) participants [4]. Utilizing EEG measure, a language-related left *hypo*frontality was found. Moreover, "in depressed patients, positive affect was negatively correlated with depression levels and positively correlated with left frontal EEG beta amplitude" [4]. That is, when processing positive affect, there was an increase in activation in the left frontal area in MDD patients; conversely, we can say that a processing of negative affect would lead to a decrease in activation in the left frontal area with the assumption that the brain processes these opposite affects in an opposite fashion in regards to brain activations. Although left frontal beta activations might not directly map left fronto-temporal language related network activations, we hope to formulate our hypothesis based on this estimation or extension of the results in this study.

# Hypothesis


From our background research, we hypothesize that exposure to emotionally positive words during a passive listening task will lead to **decreased** activation of the Default Mode Network (DMN) in healthy adults compared to exposure to neutral or emotionally negative words based on the knowledge of the association between DMN hyperactivity and depression. Given that higher DMN activations are related with depression, a condition in which one experiences difficulties in emotional regulation and usually practices more negative self talk/thoughts, we hypothesize that positive emotional stimuli will result in lower levels of DMN activation. Conversely, we expect an **increase** in activation of the language cortex in response to positive sentiment stimuli, reflecting the finding that language-related left hypofrontality is found in depressive patients with MDD and similarly extending this into how positive words might influence such brain areas.

## Sentiment Analysis

To visualize our data and for the purpose of our project, let's first apply sentiment analysis to our 627 sentence stimuli to separate them into positive, negative, or neutral sentences

```python
analyzer = SentimentIntensityAnalyzer()

# Function to get sentiment score
def get_sentiment(sentence):
    return analyzer.polarity_scores(sentence)['compound']

# Apply sentiment analysis to each sentence
stimulus['sentiment'] = stimulus['sentence'].apply(get_sentiment)
```

Save the dataframe with the recently added sentiment scores into a csv file for easier usage later on (especially when you re-visit the notebook and don't want to run the sentiment analyzer again)

```python
# Save the updated dataframe with sentiment scores
stimulus.to_csv('stimuli_with_sentiment.csv', index=False)

stimuli_with_sentiment = pd.read_csv('stimuli_with_sentiment.csv')

stimuli_with_sentiment.head()
```

Now that we have the sentiment *scores*, we also want the sentiment *labels*

```python
# Categorize sentiment scores into positive, negative, and neutral
def categorize_sentiment(score):
    if score > 0.05:
        return 'positive'
    elif score < -0.05:
        return 'negative'
    else:
        return 'neutral'

stimuli_with_sentiment['sentiment_category'] = stimuli_with_sentiment['sentiment'].apply(categorize_sentiment)
```

### Histogram of DMN Average Response to All Stimuli

```python
# Initialize a list to store the average response across all stimuli for each patient
average_response_per_patient = []

# Find the minimum number of features (voxels) across all patients
min_num_features = min(len(activations[0]) for activations in dmn_data.values())

# Iterate over each patient's data
for patient_id, activations in dmn_data.items():
    # Trim or pad the activations to ensure all have the same length
    activations = [activation[:min_num_features] for activation in activations]
    # Calculate the average response across all stimuli for the current patient
    average_response = np.mean(activations, axis=0)
    average_response_per_patient.append(average_response)

# Calculate the overall average response across patients
overall_average_response = np.mean(average_response_per_patient, axis=0)

# Plot the overall average response as a histogram
plt.hist(overall_average_response, bins=30)
plt.xlabel('DMN Activation')
plt.ylabel('Frequency')
plt.title('Histogram of DMN Network Average Response to All Stimuli')
plt.show()
```

# Data Analysis & Results

## Average DMN for all patients in separate graph based on sentiment

```python
# Extract sentence numbers for each sentiment category
pos_sentence_nums = stimuli_with_sentiment[stimuli_with_sentiment['sentiment_category'] == 'positive']['sentence_num'].values
neg_sentence_nums = stimuli_with_sentiment[stimuli_with_sentiment['sentiment_category'] == 'negative']['sentence_num'].values
neut_sentence_nums = stimuli_with_sentiment[stimuli_with_sentiment['sentiment_category'] == 'neutral']['sentence_num'].values

# Function to calculate average activations for a given set of sentence numbers
def calculate_average_activation(sentence_nums, activations):
    selected_activations = activations[sentence_nums - 1]  # Adjust for 0-based indexing
    average_activation = np.mean(selected_activations, axis=0)
    return average_activation

# Find the maximum shape among all patients
max_shape = max([activations.shape[1] for activations in dmn_data.values()])

# Function to pad activations to the maximum shape
def pad_activations(activations, max_shape):
    padded_activations = np.zeros((activations.shape[0], max_shape))
    padded_activations[:, :activations.shape[1]] = activations
    return padded_activations

# Initialize lists to hold average activations for each sentiment category
avg_activations_positive = []
avg_activations_negative = []
avg_activations_neutral = []

# Loop through each patient and calculate average activations for each sentiment category
for patient_id in dmn_data.keys():
    patient_activations = pad_activations(dmn_data[patient_id], max_shape)
    avg_activations_positive.append(calculate_average_activation(pos_sentence_nums, patient_activations))
    avg_activations_negative.append(calculate_average_activation(neg_sentence_nums, patient_activations))
    avg_activations_neutral.append(calculate_average_activation(neut_sentence_nums, patient_activations))

# Calculate the overall average activations across all patients for each sentiment category
overall_avg_activation_positive = np.mean(avg_activations_positive, axis=0)
overall_avg_activation_negative = np.mean(avg_activations_negative, axis=0)
overall_avg_activation_neutral = np.mean(avg_activations_neutral, axis=0)

# Plot the overall average activations for each sentiment category in separate graphs
plt.figure(figsize=(15, 10))

# Plot for Positive Sentences
plt.subplot(3, 1, 1)
plt.plot(overall_avg_activation_positive)
plt.xlabel('Voxel Index')
plt.ylabel('Average Activation')
plt.title('Overall Average fMRI Activation for Positive Sentences (for all patients)')

# Plot for Negative Sentences
plt.subplot(3, 1, 2)
plt.plot(overall_avg_activation_negative)
plt.xlabel('Voxel Index')
plt.ylabel('Average Activation')
plt.title('Overall Average fMRI Activation for Negative Sentences (for all patients)')

# Plot for Neutral Sentences
plt.subplot(3, 1, 3)
plt.plot(overall_avg_activation_neutral)
plt.xlabel('Voxel Index')
plt.ylabel('Average Activation')
plt.title('Overall Average fMRI Activation for Neutral Sentences (for all patients)')

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

  - neutral sentences exhibited a higher response than negative sentences

 - Although there were no significant effects while trying to encode (stimuli to brain) different sentiment valued words onto DMN or language cortex brain activations (biggest difference was 0.2 between activaitons), the language cortex demonstrated a stronger effect than the DMN. This showcases that based on our background which we used to build our hypothesis, the DMN's involvement in depression might be in response to another aspect other than the processing of emotionally charged words (Language cortex being more responsive to sentences might showcase language cortex greater role/involvement in semantic and emotional processing and potentially thus more influence in the emotional processing factor of linguistic stimuli in depression than the DMN).

## Logistic Regression:

We used **logistic regression** for decoding positive, neutral, or negative sentence stimuli (categorical variables) from brain activations.

The code uses a logistic regression model (LogisticRegression) with the multi_class='multinomial' parameter, which is appropriate for multiclass classification.

### Decoding Sentiment Categories:

The model is trained to classify brain activations into three sentiment categories: positive, neutral, or negative.
The categorize_sentiment function categorizes the sentiment scores into these three categories.
The sentiment_mapping maps these categories to numeric labels used for training the logistic regression model.

### Categorical Variables:

The sentiment categories (positive, neutral, negative) are categorical variables, and the logistic regression model is used to decode these from the brain activation data.

### Brain Activations:

The X matrix consists of brain activation data for different participants, which is used as the input features for the logistic regression model.
The code includes preprocessing steps to pad the arrays and handle NaN or infinite values in the brain activation data. 

### Interpretation of Logistic Regression
The language cortex having a higher accuracy signifies that the language cortex is better at decoding brain activations into different sentiment labeled sentences than the DMN. This suggests that the language cortex 'better processes' or more accurately predicts linguistic stimuli - it is more effective towards understanding the semantics of sentences by categorizing each sentence into its *correct* sentiment category by 92%. One could say from this result that the language cortex is more highly specialized at processing meaningul linguistic stimuli than the DMN. This is non surprising because the language cortex is known for its functions of understanding linguistic input and producing language - signifying that the language cortex is specialized to process the sentiment of a sentence. If the language cortex has higher decoder abilites than the DMN or better processes linguistic stimuli, then studying the DMN in regards to processing the *emotional content* of a word might not be the most suited way of uncovering its effect in depression (our motivation behind this research to potentially extend our results to MDD patients). If it is established that DMN hyperactivity is associated with depression but the DMN's functionality/specialty is not processing the sentiment of a word, which is what we're analyzing in this project as a potential indirect measure of its effect on depression, then this leaves room for further testing other aspects of the DMN (other than responding to the valence of a word) that might be linked more strongly to depression.
